## Exploratory Data Analysis

###  Have movie revenues and budgets changed over time? Are there certain months where movies higher earnings?

This question will be analyzed in three sections: revenue over time, budget over time, and revenue grouped by months.

First, look at how have revenues changed over time.

```python
#filter out rows with no reported revenue
df_rev = df_total[df_total['revenue_adj'] > 1] 
df_rev = df_rev.reindex()

#look at raw revenue_adj over time
plt.figure(1);
plt.figure(figsize=(15,7.5));
plt.scatter(df_rev['release_year'],df_rev['revenue_adj']);
plt.xlabel('Release Year');
plt.ylabel('Adjusted Revenue in Billions ($)');
plt.title('Adjusted Revenue over Time');

#look at yearly average over time
years = df_rev['release_year'].unique()
avg_rev_per_year = []
for year in years:
    n = df_rev[df_rev['release_year'] == year].count()['revenue_adj']
    total = sum(df_rev[df_rev['release_year'] == year]['revenue_adj'])
    avg_rev_per_year.append((year,total/n))

avg_year = []
avg_rev = []
for point in avg_rev_per_year:
    avg_year.append(point[0]), avg_rev.append(point[1])
plt.figure(2);
plt.figure(figsize=(15,7.5));
plt.scatter(avg_year,avg_rev, marker = '+', color = 'red');
plt.xlabel('Release Year');
plt.ylabel('Adjusted Revenue in Billions ($)');
plt.title('Average Adjusted Revenue over Time');

```

The first plot shows a raw scatter plot of adjusted revenue over time. It is evident that there is a large number of outliers for each year. This is impactful on the second plot, which shows the average adjusted revenue over time. Due to the low count of datapoints relative to outliers in each year, the trend before 1980 is very noisy. However, from 1980 to the present, the average is fairly stable. This indicates little change in movie revenues over time.  
  
Next budget over time will be analyzed.

```python
#filter out rows with no reported budget
df_bud = df_total[df_total['budget_adj'] > 1] 
df_bud = df_bud.reindex()

#look at raw budget_adj over time
plt.figure(1);
plt.figure(figsize=(15,7.5));
plt.scatter(df_bud['release_year'],df_bud['budget_adj']);
plt.xlabel('Release Year');
plt.ylabel('Adjusted Budget in Tens of Millions ($)');
plt.title('Adjusted Budget over Time');

#look at yearly average over time
years = df_bud['release_year'].unique()
avg_rev_per_year = []
for year in years:
    n = df_bud[df_bud['release_year'] == year].count()['budget_adj']
    total = sum(df_bud[df_bud['release_year'] == year]['budget_adj'])
    avg_rev_per_year.append((year,total/n))

avg_year = []
avg_bud = []
for point in avg_rev_per_year:
    avg_year.append(point[0]), avg_bud.append(point[1])
plt.figure(2);
plt.figure(figsize=(15,7.5));
plt.scatter(avg_year,avg_bud, marker = '+', color = 'red');
plt.xlabel('Release Year');
plt.ylabel('Adjusted Budget in Tens of Millions($)');
plt.title('Average Adjusted Budget over Time');
```

Unlike the movie revenues, which seemed to have a steady average after 1980, the budgets increased following 1970. The average budget peaked near 2000 and has fallen consistently to the present.  
  
Now for a look at whether certian months have higher averages than others.

```python
df_rev['release_month'] = df_rev['release_date'].map(lambda x: x.month)
df_mean = df_rev.groupby(df_rev['release_month']).mean()

df_mean['revenue_adj'].plot(kind = 'line', legend = True,
                            figsize=(7.5,3.75), title = 'Average Adjusted Revenue Grouped by Release Month');
ax = df_mean['budget_adj'].plot(kind = 'line', legend = True);
ax.set_ylabel('Avg Revenue/Budget ($)');
```

From the plot shown above, there certianly seems to be a correlation between release month and average revenue/budget. The two peaks occur during the summer months and the holiday season (november/december).
#### Question 1 Summary:
Average adjusted revenue from movies appears to be fairly constant over time from 1980. Despite this consistancy in revenue, adjusted budget shows an increase from 1980 to about 2000, where it starts decreasing over time. When budget and revenue were grouped by release month, there was a clear correlation between high revenue movies during the summer and holiday season. This correlation between release month and revenue was mirrored by budget.

### Are specific actors, directors, or production companies, associated with high grossing movies? Additionally, do high grossing movies receive higher populatity and higher ratings than other movies?

To begin answering the first question in this set, the dataset will be limited to non-null values of cast, directors, and production companies. The rows with 0 listed for adjusted revenue will also be removed as this is most likely an error.  
The first question refers to "high grossing movies", this will be defined as movies above 90% percentile of the adjusted revenue. The following rankings will be used:  
 - 10% and below - very low
 - 11% - 25% - low
 - 26% - 50% - medium low
 - 51% - 75% - medium high
 - 76% - 90% - high
 - Above 90% - very high

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate..

#filter dataframe by removing nulls and 0s from revenue_adj
df_cast = df_total[(df_total['cast'].notnull()) & 
                   (df_total['director'].notnull()) & 
                   (df_total['production_companies'].notnull()) &
                   (df_total['revenue_adj'] > 1.0)]
df_cast = df_cast.reset_index()

#setup qualtiles for revenue levels
rev_lim = (df_cast['revenue'].quantile(0.1),
                 df_cast['revenue'].quantile(0.25),
                 df_cast['revenue'].quantile(0.50),
                 df_cast['revenue'].quantile(0.75),
                 df_cast['revenue'].quantile(0.9))

#add revenue level column based on percentage given in description above
revenue_levels = []
for row in range(df_cast.shape[0]):
    if df_cast['revenue'][row] <= rev_lim[0]:
        revenue_levels.append('very low')
    elif (df_cast['revenue'][row] > rev_lim[0]) & (df_cast['revenue'][row] <= rev_lim[1]):
        revenue_levels.append('low')
    elif (df_cast['revenue'][row] > rev_lim[1]) & (df_cast['revenue'][row] <= rev_lim[2]):
        revenue_levels.append('medium low')
    elif (df_cast['revenue'][row] > rev_lim[2]) & (df_cast['revenue'][row] <= rev_lim[3]):
        revenue_levels.append('medium high')
    elif (df_cast['revenue'][row] > rev_lim[3]) & (df_cast['revenue'][row] <= rev_lim[4]):
        revenue_levels.append('high')
    else:
        revenue_levels.append('very high')
df_cast['revenue_level'] = revenue_levels

#create datafrome from very high revenue levels
df_cast_vh = df_cast[df_cast['revenue_level'] == 'very high'].reset_index()

def plot_bar(df, series, num = 25):
    cast_list = []
    for row in range(df.shape[0]):
        cast_list += df[series][row].split('|')
    names = Counter(cast_list) #count occurences of the unique cast members
    names = names.most_common()[:num] #25 most common cast members
    ind = pd.Index(['very high','high','medium high','medium low','low','very low'])
    fig, axes = plt.subplots(nrows = 1, ncols = 1)
    for name in names:
        ax = df_cast['revenue_level'][df_cast[series].str.contains(name[0])].value_counts()[ind].plot(kind = 'bar');
    ax.set_title(series.title().replace('_',' ') + ' Counts in Revenue Levels')
    ax.set_ylabel('Row Count')
```

The following produces two plots of most common unique cast members from two data sets. The first plot counts the occurance of the 25 most common cast members from the *whole dataset* in the revenue levels. The second plot counts the occurance of the 25 most common cast members from the *'very high' revenue level rows* in the revenue levels. In both plots, the counts are skewed right, suggesting certain actors may correlate to higher revenue movies. The second plot is skewed more to the right than the first plot. This should be expected since the cast members selected where selected from the 'very high' revenue level rows. In the second plot, the distribution of occurances of the top cast members is almost uniform in the top three revenue levels.  
  
  Overall, this data seems to suggest that certain cast members correlate to high revenue movies.

```python
plot_bar(df_cast, 'cast')
plot_bar(df_cast_vh, 'cast')
```

Like in the previous plots about cast members, the following two plots are occurances of the top 25 directors in the revenue levels. Again, the directors used in the first plot are the top 25 most common in the whole dataset, while the directors in the second plot are the 25 most common from the 'very high' revenue level rows. These two plots are more strongly skewed right than the previous plots for cast members, indicating that certain directors are more strongly correlated to 'very high' revenue movies than cast members.  
  
  This data suggests that certain directors are correlated with higher revenue movies.

```python
plot_bar(df_cast, 'director')
plot_bar(df_cast_vh, 'director')
```

Unlike the previous two parameters, cast members and directors, production companies do not correlate strongly to high revenue movies. Again, the directors used in the first plot are the top 25 most common in the whole dataset, while the directors in the second plot are the 25 most common from the 'very high' revenue level rows. In both cases, the plots are almost normal with a slight right skew.
  
  This indicates that production companies are not strongly correlated with the high revenue movies.

```python
plot_bar(df_cast, 'production_companies')
plot_bar(df_cast_vh, 'production_companies')
```

The second question of this set relates to correlation between high revenue movies and other numeric parameters in the set: popularity, runtime, vote_average, and budget_adj. The following plots show that (on average), the other numeric parameters increase in value as movie revenue increases. 

```python
df_stats = df_cast.groupby('revenue_level').mean().sort_values(by =
            'revenue_adj').drop(['index','release_year','budget','revenue'],axis = 1)
df_stats.plot(kind = 'bar', subplots = True, layout = (2,3), figsize = (20,10));
```

#### Question 2 Summary:
From the first part of this question, the analysis shows that certain actors and director correlate to higher revenue movies, but production companies do not show a string correlation to high revenue movies.  
  
From the second part the question, higher revenue movies exbihited higher popularity and vote_average (voter's rating). The higher revenue movies also had higher budgets and runtimes. Vote_average is a weaker correlation than the other parameters, this may be due to the lower count of voters at the low revenue (and less popular) movies.
  
While these correlations are evedent from the data, causation is not accessed in this analysis.
